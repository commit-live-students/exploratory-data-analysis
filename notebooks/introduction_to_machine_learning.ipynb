## Demo 2: Plotting sine and cosine waves

```python
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C,S = np.cos(X), np.sin(X)

plt.plot(X, C)
plt.plot(X, S)
plt.show()
```

Working with shapes in a plot:

```python
plt.plot(X, C, 'r--')
plt.plot(X, S, 'g--')
plt.show()

# A few examples:
# To get squares: '<color-initial>s'
# To get triangles: '<color-initial>^'
```

Controlling the minute parameters related to the plots using the `setup()` command:

```python
lines = plt.plot(X, S)
# use keyword args
plt.setp(lines, color='r', linewidth=2.0)
plt.show()
```

Using labels, titles and annotations:

```python
lines = plt.plot(X, S)
# use keyword args
plt.setp(lines, color='r', linewidth=2.0)
plt.xlabel('x')
plt.ylabel('y')
plt.title('It is a Sine wave!')

# Note: Annotation is not essential majority of the times,
# there's always a trial&error invovled in getting it right
plt.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
            arrowprops=dict(facecolor='black', shrink=0.05),
            )

plt.show()
```

## Demo 11: Q-Q plots

```python
import numpy as np
import numpy.random as random

p = random.poisson(lam=10, size=4000)
m = np.mean(p)
s = np.std(p)
n = random.normal(loc=m, scale=s, size=p.shape)

a = m-4*s
b = m+4*s

plt.figure()
plt.plot(np.sort(n), np.sort(p), 'o', color='red')
plt.plot([a,b], [a,b], 'k-')
plt.xlim(a,b)
plt.ylim(a,b)
plt.xlabel('Normal Distribution')
plt.ylabel('Poisson Distribution with $\lambda=10$')
plt.grid(True)
save_fig("q_q_plot")
plt.show()
```

## Demo 19: Exploring a synthetic dataset

* Create a `numpy` array by
    - Sample 90 points from N(15, 5)
    - Sample 10 points from N(30, 10)
    - Combine the two
    - Shuffle the list
    - Call it `age2`
* For the array, calculate
    - the mean
    - the median
    - the mode

```python
age2_1 = np.random.normal(15, 5, 90)
age2_2 = np.random.normal(30 ,10, 10)
age2 = np.concatenate((age2_1, age2_2))
np.random.shuffle(age2)
```

```python
age2_mean = np.average(age2)
age2_median = np.median(age2)
age_2_mode = stats.mode(age2)[0][0]

print("Mean:", age2_mean)
print("Median:", age2_median)
print("Mode:", age_2_mode)
```

## Demo 21: Titanic Embark Mode

* From titanic `train` dataset, extract the `embark` column as numpy array
    - name it `train_embark`
* Create frequency table
* Calculate mode

```python
train_embarked = np.array(train.Embarked)
test_embarked = np.array(test.Embarked)
```

```python
enc = LabelEncoder()
train.Embarked.fillna("S", inplace=True)
train_embarked = np.array(enc.fit_transform(train.Embarked))
test_embarked = np.array(enc.transform(test.Embarked))
unique, counts = np.unique(train_embarked, return_counts=True)
[print(enc.classes_[i], " : " ,counts[i]) for i in range(3)]
```

```python
train_embarked_mode = stats.mode(train_embarked)[0][0]
print("train_embarked_mode:", enc.classes_[train_embarked_mode])
```

```python
mean_RMSE = sqrt(mean_squared_error(test_age[~np.isnan(test_age)], [train_age_mean]*test_rows))
median_RMSE = sqrt(mean_squared_error(test_age[~np.isnan(test_age)], [train_age_median]*test_rows))
mode_accuracy = accuracy_score(test_embarked, [train_embarked_mode]*len(test_embarked))

print("{:<13} : {:.3f}".format("Mean_RMSE", mean_RMSE))
print("{:<13} : {:.3f}".format("Median_RMSE", median_RMSE))
print("{:<13} : {:.2f}%".format("Mode_accuracy", mode_accuracy*100))
```

## Demo 23: Dispersion


* For `train_age`, `age2` and `train_embark`, calculate
    - Standard deviation
    - Interquartile range (IQR)
    - Range
    - Mean absolute difference
    - (optional) Median absolute deviation (MAD)
    - Variance
    - Entropy
    
https://en.wikipedia.org/wiki/Qualitative_variation

```python
for dist in [train_age, age2]:
    print("=======")
    print("Std:", np.nanstd(dist))
    i75, i25 = np.nanpercentile(dist, [75, 25])
    print("IQR:", i75 - i25)
    mini = np.nanmin(dist)
    maxi = np.nanmax(dist)
    print("MAD:", robust.mad(dist[~np.isnan(dist)]))
    print("Range:", maxi - mini)
    print("Variance:", np.nanvar(dist))
    print("Entropy:", stats.entropy(dist[~np.isnan(dist)]))
```
