# Statistics: Introduction

## What is Statisitcs?

* The study of the methods for obtaining, organizing, analyzing, and interpreting data.


## Why Statistics? 

Principles provide a framework for 

* collecting data and the design of experiments and observational studies (Design)
* Describing and summarizing data (Description)
* Drawing inferences about populations as a whole and predicting future events (Inference)


* Statistics is the science of using data to prove a point (and hopefully forming a correct conclusion).

## Population vs Sample


* The **population** includes all objects of interest 
* The **sample** is only a portion of the population. 


* Population characteristics are known as **Parameters** 
* Sample characteristics are known as **Statistics** 


* Parameters are usually denoted using Greek letters (mu, sigma) 
* Statistics are usually denoted using Roman letters (x, s).

### Reasons why we don't work with populations. 

* They are usually large, and it is often impossible to get data for every object we're studying. 
* Sampling does not usually occur without cost, and the more items surveyed, the larger the cost.


* We compute statistics, and use them to estimate parameters. 
* The computation is known as **Descriptive Statistics** and the estimation is known as **Inferential Statistics**

## Variables and their Measurements

* Variable: Any characteristic that takes different values for different individuals in a sample or population
* Measurement scale: the type of measurements that the values of a variable take. 3 kinds: nominal, ordinal, and interval.


* Two major Types of variables: **categorical** and **quantitative**


* A categorical variable is a variable that can take on a few different values (categories) when measured. 
* Sometimes called qualitative variables.


* A quantitative variable is a variable that is measured on a numerical scale covering a large range of values.

## Quantitative Variables:

* **Discrete variables** are usually obtained by counting. 
* There are a finite or countable number of choices available with discrete data. 
* e.g. You can't have 2.63 people in the room.


* **Continuous variables** are usually obtained by measuring. 
* e.g. Length, weight, and time etc. 
* Since continuous variables are real numbers, we usually round them. 
* This implies a boundary depending on the number of decimal places. 
    * For example: 64 is really anything 63.5 <= x < 64.5. 
    * Likewise, if there are two decimal places, then 64.03 is really anything 63.025 <= x < 63.035. 
    * Boundaries always have one more decimal place than the data and end in a 5.

## Categorical Variables:

There are four levels of measurement: 

1. Nominal
- Ordinal 
- Interval
- Ratio. 


* These go from lowest level to highest level. 
* Data is classified according to the highest level which it fits. 
* Each additional level adds something the previous level didn't have.


* Nominal is the lowest level. Only names are meaningful here.
* Ordinal adds an order to the names.
* Interval adds meaningful differences
* Ratio adds a zero so that ratios are meaningful.

## Sampling

Sampling is the method used to obtain a sample of individuals/subjects from a larger population.

### Randomization in sampling:

* Some form of randomization should be used. That is chance is used to determine which individuals in the population end up in the sample.
* This ensures that the sample is representative of the population (similar make-up of individuals)


### Types of Sampling:


#### Simple random sample (SRS)

* each individual in the population has an equal chance of being chosen
* every set of n individuals has an equal chance of being
the sample chosen

#### Stratified random samples

Basic idea: sample important groups separately,
then combine these samples
1. Divide population into groups of similar
individuals, called strata
2. Choose a separate simple random sample within
each strata
3. Combine the results of the simple random
samples together to form the overall statistic,
weighting each separate stratum correctly to
mimic the population


### Sampling Bias: cautions for surveys:

Statistical bias refers to measurement or sampling errors that are systematic and pro‐
duced by the measurement or sampling process. 

An important distinction should be
made between errors due to random chance, and errors due to bias

#### Types of Bias:

1. **Selection bias**: some groups in population are
over- or under-represented in sample
(the sampling frame is limited)
2. **Non-response bias**: non-respondents may differ
in important ways from respondents
(individuals choose not to respond)
3. **Response bias**: e.g., wording of questions,
telescoping in the recall of events

### Sampling Error:

The sampling error of a statistic is the error that occurs
when we use a statistic based on a sample to predict the
value of a parameter in the population.

* Random sampling prevents bias in that we expect the
sampling error to be around zero. 
* Sometimes our statistic
will be above the parameter (positive sample error), and
sometimes it will be below it (negative sampling error).
* Eventually we will be able to predict/control the amount
of sampling error (called the margin of error). 


**Important**: Data Quality is often more impotant than data quantity. Randomly collected data can help reduce bias improve data quality.

* Quick Quiz: Would you expect this margin of error to go up or down
with a larger sample size?

### Bias and variability:

* Sampling error can be broken down into two parts:
    * Bias – the systematic error and distance of a sample
statistics from the population parameter
    *  Variability – the expected variation in the sampling
error of the sample statistic
* Bias is often thought of as a measure of validity of a
study design (reduced by using random sampling)
* Variability captures the spread in the sample statistic (
reduced by increasing sample size)
* If bias = 0 and variability is small, the values of a
statistic will be tightly clustered around the “truth”
